```python
import os

if not os.path.exists('./dcard_photo'):
    os.mkdir('./dcard_photo')

for articleObj in jsonData:
    title = articleObj['title']
    imgList = articleObj['mediaMeta']
    print(title)
    for idx, imgObj in enumerate(imgList):
        print('\t', imgObj['url'])
        with open('./dcard_photo/{}_{}.{}'.format(title, idx, imgObj['url'].split(".")[-1]), 'wb') as f:
            f.write(
                requests.get(imgObj['url'], headers=headers).content
            )
```
